```python
from flask import Flask, request
import random 
import datetime
```

```python
app= Flask(__name__)
```

Uppercase

```python
@app.route('/')
def uppercase_name():
    name= request.args.get("name")

    if not name:
        return "please provide name using ?name=yourname"

    return name.upper()

app.run( debug=True, use_reloader=False)
```

```python
app.run( debug=True, use_reloader=False)
```

```python

app.run(host= "127.0.0.1", port= 5000, debug= True, use_reloader= False)
```

Bonus: Reversing name


```python
@app.route("/reverse")
def reverse_name():
    name= request.args.get("name")

    if not name:
        return "please provide name using ?name=yourname"

    return name.reverse()

app.run( debug=True, use_reloader=False)

```

```python
app.run(host= "127.0.0.2", port= 5000, debug= True, use_reloader= False)
```
